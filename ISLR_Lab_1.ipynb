## 2.3.4 Graphics

```python
from matplotlib.pyplot import subplots
fig, ax = subplots(figsize=(8,8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x,y);
```

```python
output = subplots(figsize=(8,8))
fig = output[0]
ax = output[1]
```

```python
fig, ax = subplots(figsize=(8,8))
ax.plot(x, y, 'o')
```

```python
fig, ax  = subplots(figsize=(8, 8))
ax.scatter(x, y, marker='o')
```

```python
# np.linspace(a, b, n) -> returns a vector of n numbers starting at a and ending at b.
fig, ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)
y = x
f = np.multiply.outer(np.cos(y), 1 /(1+x**2))
ax.contour(x, y, f)
```

```python
# level을 설정해서 image의 resolution을 높임
fig, ax = subplots(figsize=(8, 8))
ax.contour(x, y, f, levels=45)
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.imshow(f) # similar to ax.contour(), but it produces a color-coded plot whose colors depend on the z value.
```

## Using Seaborn

```python
import numpy as np
def dummy_npwarn_decorator_factory():
  def npwarn_decorator(x):
    return x
  return npwarn_decorator
np._no_nep50_warning = getattr(np, '_no_nep50_warning', dummy_npwarn_decorator_factory)
import seaborn as sns
sns.relplot(x='cylinders',y='mpg', data=Auto)
```

```python
sns.catplot(x='cylinders', y='mpg', kind='swarm', data=Auto)
```
